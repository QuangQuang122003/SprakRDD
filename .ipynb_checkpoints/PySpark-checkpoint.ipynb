# Word count
B1. LOAD (file.txt)

B2. Transformations[ FLATMAP(), MAP(), REDUCEBYKEY()]

B3. Action ( COLLECT() )

```python
rdd1 = spark.sparkContext.textFile("/Users/mac/Documents/data.txt")
```

```python
#Ktra xem doc dc file du lieu chua?
rdd1.collect()
```

```python
rdd2 = rdd1.flatMap(lambda x : x.split(" "))
```

```python
rdd2.collect()
```

```python
rdd3 = rdd2.map(lambda word : (word,1))
```

```python
rdd3.collect()
```

```python
rdd4 = rdd3.reduceByKey(lambda x,y : x+y)
```

```python
rdd4.collect()
```
