# Spark RDD Thực Hành

```python
rdd = spark.sparkContext.textFile("/Users/mac/Documents/SprakRDD/part-m-00000.txt")
```

```python
rdd.take(10)
```

```python
# So luong orders tuong ung voi moi status
```

```python
order = rdd.map(lambda x : (x.split(",")[3], 1))
order.take(10)
```

```python
status = order.reduceByKey(lambda x,y : x+y)
status.collect()
```

```python
sorted_order = status.sortBy(lambda x : x[1], False)
sorted_order.collect()
```

```python
# Khach hang nao co status order == CLOSE nhieu nhat
```

```python
filter_cus = rdd.filter(lambda x : x.split(",")[3] == 'CLOSED')
filter_cus.take(10)
```

```python
map_cus = filter_cus.map(lambda x : (x.split(",")[2],1))
map_cus.take(10)
```

```python
reduce_cus = map_cus.reduceByKey(lambda x,y : x+y)
reduce_cus.take(10)
```

```python
sort_cus = reduce_cus.sortBy(lambda x : x[1], False)
sort_cus.take(1)
```
